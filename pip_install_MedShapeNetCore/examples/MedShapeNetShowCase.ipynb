```python
from MedShapeNetCore.MedShapeNetCore import MyDict,MSNLoader,MSNVisualizer,MSNSaver,MSNTransformer
msn_loader=MSNLoader()
msn_vis=MSNVisualizer()
```

```python
ThoracicAorta_Saitta_data=msn_loader.load('ThoracicAorta_Saitta')
```

```python
CoronaryArteries_data=msn_loader.load('CoronaryArteries')
```

```python
points_ThoracicAorta=ThoracicAorta_Saitta_data['point']
print(len(points_ThoracicAorta))

points_CoronaryArteries=CoronaryArteries_data['point']
print(len(points_CoronaryArteries))
```

```python
%matplotlib notebook
```

```python
msn_vis.plot_point(points_ThoracicAorta[0][:,0],points_ThoracicAorta[0][:,1],points_ThoracicAorta[0][:,2])
```

```python
msn_vis.plot_point(points_CoronaryArteries[0][:,0],points_CoronaryArteries[0][:,1],points_CoronaryArteries[0][:,2])
```
